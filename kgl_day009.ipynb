```python
import pandas as pd
```

```python
file_path='D:/lh_data/melb_data.csv'
melbourne_data=pd.read_csv(file_path)
melbourne_data
```

```python
for i in melbourne_data.columns:                   # percent display of missing values
    null_rate = melbourne_data[i].isna().sum() / len(melbourne_data) * 100 #
    if null_rate > 0 :                             #
        print("{} null rate: {}%".format(i,round(null_rate,2)))	
```

```python
melbourne_data.columns
```

```python
melbourne_data = melbourne_data.dropna(axis=0)
```

```python
y=melbourne_data['Price'].copy()
y
```

```python
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
X=melbourne_data[melbourne_features]
X
```

```python
X.describe()
```

```python
X.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
melbourne_model = DecisionTreeRegressor(random_state=1)
# Fit model
melbourne_model.fit(X, y)
```

```python
print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are")
print(melbourne_model.predict(X.head()))
```
